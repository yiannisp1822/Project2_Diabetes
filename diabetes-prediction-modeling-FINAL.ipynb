### Feature Grouping (using functions)

#### Age
1. Age 18 to 24
2. Age 25 to 29
3. Age 30 to 34
4. Age 35 to 39
5. Age 40 to 44
6. Age 45 to 49
7. Age 50 to 54
8. Age 55 to 59
9. Age 60 to 64
10. Age 65 to 69
11. Age 70 to 74
12. Age 75 to 79
13. Age 80 or older

#### split every three groups
#### 1 = 18-34 (1-3), 2 = 35-49 (4-6), 3 = 50-64 (7-9), 4 = 65-79 (10-12), 5 = 80+ (13) 

```python
diabetes_df['Age'].value_counts()
```

```python
def age_replace(x):
    if x >= 1 and x <= 3:
        return "Age 18-34"
    elif x > 3 and x <= 6:
        return "Age 35-49"
    elif x > 6 and x <= 9:
        return "Age 50-64"
    elif x > 9 and x <= 12:
        return "Age 65-79"
    elif x > 12:
        return "Age 80+"
```

```python
diabetes_df['Age'] = diabetes_df['Age'].apply(age_replace)
diabetes_df['Age'].value_counts()
```

#### EDUCATION
1. Never attended school or only kindergarten
2. Grades 1 through 8 (Elementary)
3. Grades 9 through 11 (Some high school)
4. Grade 12 or GED (High school graduate)
5. College 1 year to 3 years (Some college or technical school)
6. College 4 years or more (College graduate)

#### split between : higher vs non-higher edu, 5-6 higher, 1-4 lower

```python
def edu_replace(x):
    if x >= 1 and x <= 4:
        return "Lower Education"
    elif x > 4:
        return "Higher Education"
```

```python
diabetes_df['Education'] = diabetes_df['Education'].apply(edu_replace)
diabetes_df['Education'].value_counts()
```

### INCOME
1. Less than $10,000
2. Less than $15,000 ($10,000 to less than $15,000)
3. Less than $20,000 ($15,000 to less than $20,000)
4. Less than $25,000 ($20,000 to less than $25,000)
5. Less than $35,000 ($25,000 to less than $35,000)
6. Less than $50,000 ($35,000 to less than $50,000)
7. Less than $75,000 ($50,000 to less than $75,000)
8. $75,000 or more

#### group 1-3, 4-7, 8 by itself

```python
diabetes_df['Income'].value_counts()
```

```python
def income_replace(x):
    if x >= 1 and x <= 3:
        return "Less than $20,000"
    elif x > 3 and x <= 7:
        return "Between $20,000 and $75,000"
    elif x > 7:
        return "More than $75,000"
```

```python
diabetes_df['Income'] = diabetes_df['Income'].apply(income_replace)
diabetes_df['Income'].value_counts()
```

### BMI

```python
def BMI_classification(x):
    if x < 18.5:
        return "Underweight"
    elif x > 18.5 and x <=24.9:
        return "Normal"
    elif x > 24.9 and x <= 29.9:
        return "Overweight"
    elif x > 29.9 and x <= 34.9:
        return "Obesity 1"
    elif x > 34.9 and x <= 39.9:
        return "Obesity 2"
    elif x > 39.9:
        return "Obesity 3"
```

```python
diabetes_df['BMI'] = diabetes_df['BMI'].apply(BMI_classification)
diabetes_df['BMI'].value_counts()
```

### Encoding Categorical Features

#### BMI

```python
# create encoder and fit
encode_BMI = OrdinalEncoder(categories=[['Underweight', 'Normal', 'Overweight', 'Obesity 1', 'Obesity 2', 'Obesity 3']], handle_unknown='use_encoded_value', unknown_value=-1)
encode_BMI.fit(diabetes_df['BMI'].values.reshape(-1,1))
encode_BMI = encode_BMI.transform(diabetes_df['BMI'].values.reshape(-1, 1))


# create the df
encode_BMI_df = pd.DataFrame(encode_BMI, columns=['BMI'])
encode_BMI_df.head()
```

#### Income

```python
# create encoder and fit
encode_income = OrdinalEncoder(categories=[['Less than $20,000','Between $20,000 and $75,000','More than $75,000']], handle_unknown='use_encoded_value', unknown_value=-1)
encode_income.fit(diabetes_df['Income'].values.reshape(-1,1))
encode_income = encode_income.transform(diabetes_df['Income'].values.reshape(-1, 1))

# create the df
encode_income_df = pd.DataFrame(encode_income, columns=['Income'])
encode_income_df.head()
```

#### Education

```python
# create encoder and fit 
encode_educ = OrdinalEncoder(categories=[['Lower Education', 'Higher Education']], handle_unknown='use_encoded_value', unknown_value=-1)
encode_educ.fit(diabetes_df['Education'].values.reshape(-1,1))
encode_educ = encode_educ.transform(diabetes_df['Education'].values.reshape(-1, 1))

# create the df
encode_educ_df = pd.DataFrame(encode_educ, columns=['Education'])

encode_educ_df.head
```

#### Age

```python
# create encoder and fit 
encode_age = OrdinalEncoder(categories=[['Age 18-34', 'Age 35-49', 'Age 50-64', 'Age 65-79']], handle_unknown='use_encoded_value', unknown_value=-1)
encode_age.fit(diabetes_df['Age'].values.reshape(-1,1))
encode_age = encode_age.transform(diabetes_df['Age'].values.reshape(-1, 1))

# create the df
encode_age_df = pd.DataFrame(encode_age, columns=['Age'])

encode_age_df.head()
```

#### Group all encoded_dfs with the main df

```python
# create a copy of the df without the unencoded columns
diabetes_df_unencoded = diabetes_df.copy()
diabetes_df_unencoded = diabetes_df_unencoded.drop(columns=['Age', 'Education', 'Income', 'BMI'])

# add the encoded columns
diabetes_df_encoded = pd.concat([diabetes_df_unencoded, encode_BMI_df, encode_income_df, encode_educ_df, encode_age_df], axis=1)

# the results
diabetes_df_encoded.head()
```

```python
#splitting the data
X = diabetes_df_encoded.drop(columns='Diabetes_binary')                        
y = diabetes_df_encoded['Diabetes_binary']
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

```

```python
# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler()
X_scaler.fit(X_train)
```

```python
# Transform the training and testing data by using the X_scaler model
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Loop through different k values to find which has the highest accuracy.
# Note: We use only odd numbers because we don't want any ties.
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
# Plot the results
plt.plot(range(1, 20, 2), train_scores, marker='o', label="training scores")
plt.plot(range(1, 20, 2), test_scores, marker="x", label="testing scores")
plt.xlabel("k neighbors")
plt.ylabel("accuracy score")
plt.legend()
plt.show()
```

```python
logistic_regression_model = LogisticRegression(random_state=1)
logistic_regression_model.fit(X_train_scaled, y_train)

#make and save testing predictions with the saved logistic regression model using the test data
predictions = logistic_regression_model.predict(X_test_scaled)

# Review the predictions

predictions
```

```python
accuracy = accuracy_score(y_test, predictions)

accuracy
```

```python
# Create the decision tree classifier instance
tree_model = tree.DecisionTreeClassifier()
```

```python
#fit the model
tree_model = tree_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = tree_model.predict(X_test_scaled)
```

```python
# Calculate the accuracy score
acc_score = accuracy_score(y_test, predictions)

print(f"Accuracy Score : {acc_score}")
```

```python
# Train the Random Forest model
clf = RandomForestClassifier(random_state=1, n_estimators=1000).fit(X_train_scaled, y_train)

# Evaluate the model
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```
